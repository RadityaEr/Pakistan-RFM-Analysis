```python
# Handle Missing value
data_drop = dataset.dropna()
data_drop.info()
```

Noting
Created at = Date of Order

Ada Negatif Grand Total dan Discount Amount

```python
# Membuang Kolom yang Tidak Relevan atau Redundant
columns_to_drop = ['Working Date', 'BI Status', 'Year','Month', 'M-Y', 'FY']
dataset = dataset.drop(columns=columns_to_drop)
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
# Membuang Rows dimana 'grand_total' atayu 'discount_amount' negative
data_clean = dataset[(dataset['grand_total'] >= 0) & (dataset['discount_amount'] >= 0)]
```

```python
data_clean.describe()
```

```python
data_clean.info()
```

```python
# Convert 'created_at' ke datetime 
data_clean.loc[:, 'created_at'] = pd.to_datetime(data_clean['created_at'], errors='coerce')

# Cek kalau ada tanggal di masa depan (salah) 
invalid_future_dates = data_clean[data_clean['created_at'] > pd.Timestamp('now')]
print(invalid_future_dates)
```

```python
# Convert 'Cust_ID' ke Text 
data_clean.loc[:, 'Customer ID'] = pd.to_s(data_clean['Customer ID'], errors='coerce')
```

```python
data_clean.head()
```

```python
last_transaction_date = data_clean['created_at'].max()
last_transaction_date
```

```python
# Extract CSV

data_clean.to_csv('Clean_Pakistan_fix.csv', index=False)
```
